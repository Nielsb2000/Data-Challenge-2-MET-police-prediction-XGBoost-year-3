```python
import pandas as pd
import numpy as np
```

```python
df_deprivaiton = pd.read_csv("data/London wards id2019 summary measures.csv")
df_house_prices = pd.read_csv("data/land-registry-house-prices-ward.csv")
df_crimes = pd.read_parquet("data/all_crimes.parquet")
```

```python
df_crimes
```

```python
df_deprivaiton.columns
```

```python
##Rename deprivation and houseprice ward name columns and merge them
df_deprivaiton = df_deprivaiton.rename(columns={'Ward Name':'Ward_name'})
df_house_quality = pd.merge(df_house_prices, df_deprivaiton, on="Ward_name", how="inner").rename(columns={'Ward_name':'ward'})
```

```python
#missing values check
df_house_quality = df_house_quality.groupby('ward').median()
```

```python
len(df_crimes['ward'].unique())
```

```python
#Merge with df_crimes
merged_df = pd.merge(df_crimes,df_house_quality, on='ward', how='left')
```

```python

merged_df.to_parquet(r"data\crime_plus_demographics.parquet")
```
